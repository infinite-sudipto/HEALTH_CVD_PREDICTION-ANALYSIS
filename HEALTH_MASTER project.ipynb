# 3.Study the occurrence of CVD across Age.

```python
df.head(2)
```

```python
cvd_subset = df[['age','target']]



w_cvd = cvd_subset[cvd_subset['target'] == 1]
wo_cvd = cvd_subset[cvd_subset['target']==0]


plt.figure(figsize = (12,8))
sns.histplot(w_cvd['age'],kde = True,color = 'purple',label = 'with_CVD',edgecolor = 'black' )
sns.histplot(wo_cvd['age'],kde = True,color = 'orangered',label = 'without_CVD',edgecolor = 'black')


plt.title('DISTRIBUTION OF CVD PATIENTS OVER AGE',weight = 'bold')

plt.xlabel('AGE',weight = 'bold')
plt.ylabel('FREQUENCY',weight = 'bold')
plt.legend(loc = 'upper right')
plt.xticks(weight = 'bold')
plt.yticks(weight = 'bold')
plt.show()
```

# CONCLUSION : 
### The frequency of patients with CVD increases sharply from age 50, peaks at around age 60, and then decreases.
### he frequency of patients without CVD increases gradually from age 30, peaks at around age 60, and then decreases.
### The peak frequency for patients without CVD is higher than that for patients with CVD.

```python
x = df[['trestbps']]
y = df['target']
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_train=lr.predict(x_train)
y_pred_train
```

```python
from sklearn import metrics
print("r2 :",metrics.r2_score(y_train,y_pred_train))
```

# 2. Build a baseline model to predict using a Logistic Regression and explore the results.

```python
df.head(2)
```

```python
x  = df.drop(columns = 'target')
```

```python
y = df['target']
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test = tts(x,y,random_state = 45,test_size = 0.3)
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_sc = scaler.fit_transform(x_train)
x_test_sc = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train_sc,y_train)
```

# Y_PRED_TRAIN

```python
y_pred_train = lr.predict(x_train_sc)
y_pred_train
```

```python
from sklearn import metrics
print('acc_check :',(metrics.accuracy_score(y_train,y_pred_train)*100).round(2))
```

```python
pred_df = pd.DataFrame()
pred_df['actual'] = y_train
pred_df['predicted'] = y_pred_train
pred_df
```

# Y_PRED_TEST

```python
y_pred_test = lr.predict(x_test_sc)
y_pred_test
```

```python
pred_df2 = pd.DataFrame()
pred_df2['actual'] = y_test
pred_df2['predicted'] = y_pred_test
pred_df2
```

```python
from sklearn import metrics
print('acc_chk2 :',(metrics.accuracy_score(y_test,y_pred_test)*100).round(2))
```

# CONFUSION_MATRIX (Y_TRAIN,Y_PRED_TRAIN)

```python
cm1 = metrics.confusion_matrix(y_train,y_pred_train)
cm_df = pd.DataFrame(cm1,index= [0,1],columns = [0,1])
cm_df
```

# CLASSIFICATION_REPORT (TRAIN)

```python
from sklearn.metrics import classification_report
c_report = classification_report(y_train,y_pred_train,target_names = ['CVD YES','CVD NO'])
print(c_report)
```

# CLASSIFICATION_MATRIX(Y_TEST,Y_PRED_TEST)

```python
cm2 = metrics.confusion_matrix(y_test,y_pred_test)
cm_df2 = pd.DataFrame(cm2,columns = [0,1],index = [0,1])
cm_df2
```

# CLASSIFICATION_REPORT (TEST)

```python
from sklearn.metrics import classification_report
c_report2 = classification_report(y_test,y_pred_test,target_names = ['CVD YES','CVD NO'])
print(c_report2)
```
